# Correlation Statistics with Actual Ratings

```r
# convert ratings to matrix
rater_sd <- apply(rater_ratings, 2, sd)
rater_var <- apply(rater_ratings, 2, var)
rater_ranks <- apply(rater_ratings, 1, rank,ties.method="average")
kappam.light(rater_ratings)
kripp.alpha(t(rater_ratings), method='ordinal')
kripp.alpha(rater_ranks, method='ordinal')
kendall(rater_ratings)
kendall(rater_ratings, correct=TRUE)
cor(rater_ratings, method='spearman')
```

```r
# spearmans doesn't assume normal distribution - good bc our data is right skewed
cor(rater_ratings, method='spearman')
cor(t(rater_ranks), method='spearman')
# probably can't use this one? see if this function can break ties
meanrho(rater_ratings, fisher=FALSE)
```

# Shuffled Ratings

```r
# shuffle labels
# set seed for reproducibility
set.seed(58901)
ratings_rand <- sample(rater_ratings)
# shuffled matrix
ratings_rand <- matrix(ratings_rand, nrow = 70,ncol = 20)
colnames(ratings_rand) <- colnames(rater_ratings)
# Check that shuffled matrix is different
all(ratings_rand == rater_ratings)
ratings_rand
```

```r
apply(ratings_rand, 2, sd)
apply(ratings_rand, 2, var)
ranks_rand <- apply(ratings_rand, 1, rank,ties.method="average")
kappam.light(ratings_rand)
kripp.alpha(t(ratings_rand), method='ordinal')
kripp.alpha(ranks_rand, method='ordinal')
kendall(ratings_rand)
kendall(ratings_rand, correct=TRUE)
cor(ratings_rand, method='spearman')
```

# Charts

```r
keep_ratings = c(1,2,3,4,5)

tone1_rows <- subset(data, Target.Tone == 1)
tone1_rows <- tone1_rows[, c(seq(5,24))]
tone1_rating_freq <- table(c(data.matrix(tone1_rows)))
tone1_rating_sub <- tone1_rating_freq[keep_ratings]
barplot(tone1_rating_sub, main="Tone1 Ratings", xlab="Ratings")

tone2_rows <- subset(data, Target.Tone == 2)
tone2_rows <- tone2_rows[, c(seq(5,24))]
tone2_rating_freq <- table(c(data.matrix(tone2_rows)))
tone2_rating_sub <- tone2_rating_freq[keep_ratings]
barplot(tone2_rating_sub, main="Tone2 Ratings", xlab="Ratings")

tone3_rows <- subset(data, Target.Tone == 3)
tone3_rows <- tone3_rows[, c(seq(5,24))]
tone3_rating_freq <- table(c(data.matrix(tone3_rows)))
tone3_rating_sub <- tone3_rating_freq[keep_ratings]
barplot(tone3_rating_sub, main="Tone3 Ratings", xlab="Ratings")

tone4_rows <- subset(data, Target.Tone == 4)
tone4_rows <- tone4_rows[, c(seq(5,24))]
tone4_rating_freq <- table(c(data.matrix(tone4_rows)))
tone4_rating_sub <- tone4_rating_freq[keep_ratings]
barplot(tone4_rating_sub, main="Tone4 Ratings", xlab="Ratings")
```

```r
tones <- c('Tone1', 'Tone2', 'Tone3', 'Tone4')

for(i in 1:6) {
  # i-th element of `u1` squared into `i`-th position of `usq`
  rating_temp <- data.frame(tone1_rating_freq[i], tone2_rating_freq[i], tone3_rating_freq[i], tone4_rating_freq[i])
  colnames(rating_temp) <- tones
  barplot(as.matrix(rating_temp), main= paste("Rating", i))
}
```
